```python
##### Sampling the data  --- 
from sklearn.model_selection import train_test_split
```

```python
X = model_data['R&D Spend']
y = model_data['Profit']
```

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

```python
X_train = X_train.values.reshape(-1,1)
y_train = y_train.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
from math import sqrt
```

```python
model = LinearRegression()
model_fit = model.fit(X_train,y_train)
```

```python
model_fit.coef_
```

```python
model_fit.intercept_
```

```python
X_test = X_test.values.reshape(-1,1)
```

```python
pred = model_fit.predict(X_test)
pred
```

```python
df_pred = pd.DataFrame(pred,columns = ['R&D Spend Pred profit'])
```

```python
df_pred['Actual'] = y_test.values
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
r2_score(df_pred['Actual'],df_pred['R&D Spend Pred profit'])
```

```python
sqrt(mean_squared_error(df_pred['Actual'],df_pred['R&D Spend Pred profit']))
```
