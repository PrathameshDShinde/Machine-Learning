```python
###### Importing of Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
##### Getting the data ---
delivery_time = pd.read_csv("C:/Data Science/Oct 2023/29 Oct 2023/Project - 1&2_ Simple_Reg/Project - 1&2_ Simple_Reg/delivery_time.csv")
print(delivery_time.head())
```

```python
##### Getting the data ---
df = delivery_time
print(df.head())
print(df.info())
print(df.shape)
```

```python
#### Preprocessing and Feature Enginneering
df1 = df.copy()
print(df1.head())
```

```python
#### Exploratory Data Analysis
df1['Delivery Time'].value_counts()
```

```python
df1['Sorting Time'].value_counts()
```

```python
plt.figure(figsize=(12,4))
plt.title('Delivery_Time against Sorting_Time')
ax = sns.lineplot(x = 'Delivery Time',y = 'Sorting Time', data = df1)
plt.savefig("Delivery_Time against Sorting_Time.png")
plt.show()
```

### Inferance :- The graph shows that Delivery Time increases by Sorting Time

```python
plt.figure(figsize=(12,4))
plt.title('Delivery_Time by Sorting_Time')
sns.scatterplot(x = 'Delivery Time',y = 'Sorting Time', data = df1)
plt.savefig("Delivery_Time against Sorting_Time1.png")
plt.show()
```

### The graph shows that more time takes for sorting makes delay the delivery.

```python
plt.figure(figsize=(12,4))
plt.title('Number of Deliveries against Sorting Time')
ax = sns.barplot(x = 'Delivery Time',y = 'Sorting Time', data = df1)
for bars in ax.containers:
    ax.bar_label(bars)
plt.savefig("Delivery_Time against Sorting_Time11.png")
plt.show()
```

### Inference - The graphs shows that Highest sorting Time is 10.

```python
sns.displot(x = df1['Delivery Time'],kde = True)
print(df1['Delivery Time'].skew())
print(df1['Delivery Time'].describe())
```

```python
sns.displot(x = df1['Sorting Time'],kde = True)
print(df1['Sorting Time'].skew())
print(df1['Sorting Time'].describe())
```

```python
X = df1['Delivery Time'].values
X = X.reshape(-1,1)
print(X)
```

```python
y= df1['Sorting Time'].values
y = y.reshape(-1,1)
print(y)
```

```python
#### Importing Scikit Learn for Algorithms
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
print(model)
```

```python
#### Training the Model/System - By Fitting my data
model_fit = model.fit(X,y)
print(model_fit)
```

```python
model_fit.coef_
```

```python
model_fit.intercept_
```

```python
predict = model_fit.predict(X)
print(predict)
```
