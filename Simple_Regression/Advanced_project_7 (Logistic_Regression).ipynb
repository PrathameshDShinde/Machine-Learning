```python
#### Importing Liabraries
import pandas as pd
import datetime
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mlp
import matplotlib.ticker as ticker
%matplotlib inline
mlp.style.use('ggplot')
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
#### packages for Model Diagonistics
from sklearn.metrics import accuracy_score,confusion_matrix,ConfusionMatrixDisplay,precision_score,classification_report


#### Getting the data
def read_data():
    try:
        data = pd.read_csv("C:/Data Science/Nov 2023/19 Nov 2023/Logistic Regresssion_Project - 7/4. Logistic Regresssion/bank-full.csv")
        return data
    except Exception as e:
        print("Data Path Not Found and Error is in the Read Data Function")

#### Calling the read data function
df = read_data()
# print(df.head())

# #### Preprocessing and feature engineering
df1 = df.copy()
# print(df1.head())
# print(df1.info())
# print(df1.shape)

# print(df1['y'].value_counts())

# df1.isnull().sum()

# print(df1.describe())

# print(df1.describe(include = 'object'))

# df1.select_dtypes(include = 'number').columns


# for i in ['age', 'balance', 'day', 'duration', 'campaign', 'pdays', 'previous']:
#     sns.boxplot(data = df1,x = i)
#     plt.show()

col = ['age', 'balance', 'day', 'duration', 'campaign', 'pdays', 'previous']

def wisker(col) :
    Q1,Q3 = np.percentile(col,[25,75])
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return lower,upper

for i in col:
    lower,upper = wisker(df1[i])
    df1[i] = np.where(df1[i]<lower,lower,df[i])
    df1[i] = np.where(df1[i]>upper,upper,df1[i])

# for i in ['age', 'balance', 'day', 'duration', 'campaign', 'pdays', 'previous']:
#     sns.boxplot(data = df1,x = i)
#     plt.show()

#### Histogram to understand the distribution
# import warnings
# warnings.filterwarnings("ignore")
# plt.figure(figsize=(14,4))
# for i in df.select_dtypes(include="number").columns:
#     sns.histplot(data = df1,x = i)
#     plt.show()

# import warnings
# warnings.filterwarnings("ignore")
# plt.figure(figsize=(14,4))
# for i in df1.select_dtypes(include="object").columns:
#     sns.histplot(data = df1,x = i)
#     plt.show()

# df1.select_dtypes(include='object').columns

# #### scatter plot to understand the relationship
# for i in ['age', 'job', 'marital', 'education', 'default', 'balance', 'housing',
#        'loan', 'contact', 'day', 'month', 'duration', 'campaign', 'pdays',
#        'previous', 'poutcome']:
#     sns.scatterplot(data = df1,x = i,y = 'y')
#     plt.show()

df1 = pd.get_dummies(df1,drop_first=True)

def data_prep(df1,t):
    X = df1[['age', 'balance', 'day', 'duration', 'campaign', 'pdays', 'previous',
       'job_blue-collar', 'job_entrepreneur', 'job_housemaid',
       'job_management', 'job_retired', 'job_self-employed', 'job_services',
       'job_student', 'job_technician', 'job_unemployed', 'job_unknown',
       'marital_married', 'marital_single', 'education_secondary',
       'education_tertiary', 'education_unknown', 'default_yes', 'housing_yes',
       'loan_yes', 'contact_telephone', 'contact_unknown', 'month_aug',
       'month_dec', 'month_feb', 'month_jan', 'month_jul', 'month_jun',
       'month_mar', 'month_may', 'month_nov', 'month_oct', 'month_sep',
       'poutcome_other', 'poutcome_success', 'poutcome_unknown']]
    y = df1['y_yes']
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=t,random_state=365,stratify=y)
    return X_train,X_test,y_train,y_test

def lg_model(X_train,y_train,X_test):
    lg_model = LogisticRegression()
    lg_model_fit = lg_model.fit(X_train,y_train)
    pred = lg_model_fit.predict(X_test)
    return pred

def model_log_diag(y_test,pred):
    acc = accuracy_score(y_test,pred)
    prc = precision_score(y_test,pred)
    clf = classification_report(y_test,pred)
    conf = confusion_matrix(y_test,pred)
    return acc,prc,clf,conf

def main_function():
    df1
    X_train,X_test,y_train,y_test = data_prep(df1,0.8)
    pred = lg_model(X_train,y_train,X_test)
    cc, prc,clf,conf = model_log_diag(y_test,pred)
    return clf

results = main_function()
print(results)
```
